```python
import pandas as pd
from sklearn.model_selection import train_test_split


df = pd.read_csv("iris.csv")
(train_set, test_set) = train_test_split(
    df.values, train_size=0.7, random_state=13)

def classify_iris(sl, sw, pl, pw):
    if sl >= 4 and sl <= 6 and sw >= 2 and sw <= 4 and pl >= 1 and pl <= 2 and pw >= 0 and pw <= 1:
        return("setosa")
    elif sl >= 5 and sl <= 7 and sw >= 2 and sw <= 3.4 and pl >= 3 and pl <= 5 and pw >= 1 and pw <= 2:
        return("versicolor")
    elif sl >= 5 and sl <= 8 and sw >= 2 and sw <= 4 and pl >= 4.5 and pl <= 7 and pw >= 1.4 and pw <= 2.5:
        return("virginica")
    else:
        return("nieznany gatunek")
    
good_predictions = 0
len = test_set.shape[0]
for i in range(len):
    sl = test_set[i, 0]
    sw = test_set[i, 1]
    pl = test_set[i, 2]
    pw = test_set[i, 3]
    if classify_iris(sl,sw,pl,pw) == test_set[i, 4]:
        good_predictions = good_predictions + 1
print(good_predictions)
print(good_predictions/len*100, "%")

print(train_set)

averagesl = []
averagesw = []
averagepl = []
averagepw = []
for i in range(len):
    if train_set[i, 4] == "setosa":
        averagesl.append(train_set[i, 0])
        averagesw.append(train_set[i, 1])
        averagepl.append(train_set[i, 2])
        averagepw.append(train_set[i, 3])

data = [
    min(averagesl), max(averagesl),
    min(averagesw), max(averagesw),
    min(averagepl), max(averagepl),
    min(averagepw), max(averagepw)
]

```

```python
import pandas as pd
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

df = pd.read_csv("iris.csv")

train_set, test_set = train_test_split(df, train_size=0.7, random_state=13)

train_input = train_set.iloc[:, :-1]
train_class = train_set.iloc[:, -1]
test_input = test_set.iloc[:, :-1]
test_class = test_set.iloc[:, -1]

print("Training set:")
print(train_set)
print("Test set:")
print(test_set)

classifier = DecisionTreeClassifier()
classifier.fit(train_input, train_class)
tree.plot_tree(classifier.fit(train_input, train_class))

accuracy = classifier.score(test_input, test_class)
print(f"Dokładność klasyfikatora: {accuracy * 100:.2f}%")

predicted_class = classifier.predict(test_input)

confusion_mat = confusion_matrix(test_class, predicted_class)
print("Macierz błędów:")
print(confusion_mat)
#trening - algorytm cwiczy
#testowy - algorytm sprawdza efektywnosc
#po skosie - prawidlowo zgadniete
# reszta źle zgadniete
```

```python
import pandas as pd 
from sklearn.model_selection import train_test_split 
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

 
df = pd.read_csv("iris.csv") 
 
(train_set, test_set) = train_test_split(df.values, train_size=0.7, 
random_state=278784)

train_inputs = train_set[:, 0:4]
train_classes = train_set[:, 4]
test_inputs = test_set[:, 0:4]
test_classes = test_set[:, 4]


for k in [3, 5, 11]:
    #klasyfikator najblizszych sasiadow
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(train_inputs, train_classes)
    knn_pred = knn.predict(test_inputs)
    knn_acc = knn.score(test_inputs, test_classes)
    print(knn_acc)
    knn_cm = confusion_matrix(test_classes, knn_pred)
    print("KNN, k=", k)
    print("Poprawnosc:", knn_acc)
    print("Maciez błedu:\n", knn_cm, "\n\n")
```
