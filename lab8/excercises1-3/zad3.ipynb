```python
import time
import snscrape.modules.twitter as sntwitter

topic = "#covid19"

tweets = []

for i, tweet in enumerate(sntwitter.TwitterSearchScraper(f"{topic}").get_items()):
    tweets.append(tweet.content)
    time.sleep(1)  # OpÃ³Åºnienie 1 sekundy miÄ™dzy Å¼Ä…daniami

    if i >= 100:
        break
```

```python
for tweet in tweets:
    print(tweet + "\n")
```

```python

start_date = "2023-01-01"
end_date = "2023-05-01"

topic = "#covid19"

tweets = []

for i, tweet in enumerate(sntwitter.TwitterSearchScraper(f"{topic} since:{start_date} until:{end_date}").get_items()):
    tweets.append(tweet.content)
    time.sleep(1) 

    if i >= 50:
        break
```

```python
print("Tweets: ")
for tweet in tweets:
    print(tweet + "\n")
```
